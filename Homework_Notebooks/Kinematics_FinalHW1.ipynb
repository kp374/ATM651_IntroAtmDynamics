### Background
**Kinematics** is not dynamics, it is just a *description* of how 2D vector fields vary in 2D. 

* Assume LINEAR dependence of u,v on x,y 
    * This involves 4 coefficients $c_1, c_2, c_3, c_4$
* Also, add a constant (not variying with x,y) vector field U,V 

$$ u = U + c_1 x  +c_2 y$$
$$ v = V + c_3 x  +c_4 y$$

## Exercise 1 Answer:
* $ div = \frac{\partial u}{\partial x} + \frac{\partial v}{\partial y}= \frac{\partial u}{\partial x} (U + c_1x + c_2y) + \frac{\partial v}{\partial y} (V + c_3x + c_4y) = (0 + c_1 + 0) + (0 + 0 + c_4) = c_1 + c_4$

* $ \zeta_{rel} = \frac{\partial ?}{\partial ?} - \frac{\partial ?}{\partial ?} = \frac{\partial v}{\partial x} - \frac{\partial u}{\partial y} = c_3 - c_2 = \frac{\partial v}{\partial x} (V + c_3x + c_4y) - \frac{\partial u}{\partial y} (U + c_1x + c_2y) = (0 + c_3 + 0) - (0 + 0 + c_2) = c_3 - c_2$

* $ def1 = \frac{\partial ?}{\partial ?} + \frac{\partial ?}{\partial ?} = \frac{\partial u}{\partial x} - \frac{\partial v}{\partial y}= \frac{\partial u}{\partial x} (U + c_1x + c_2y) - \frac{\partial v}{\partial y} (V + c_3x + c_4y) = (0 + c_1 + 0) - (0 + 0 + c_4) = c_1 - c_4 $

* $ def2 = \frac{\partial ?}{\partial ?} + \frac{\partial ?}{\partial ?} = \frac{\partial v}{\partial x}+ \frac{\partial u}{\partial y} = \frac{\partial v}{\partial x} (V + c_3x + c_4y) + \frac{\partial u}{\partial y} (U + c_1x + c_2y) = c_3 + c_2 $

## Exercise 2: Create and display the basic fields
* First, execute a code cell setting up a coordinate system (x,y) from -10 to 10
* Next, execute the cell defining the basic "pure" div, vort, def1, def2, and U,V fields 
* Finally, display them. The first one has been done for you. You complete the work. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up a coordinate grid centered on 0,0
n = 10
x,y = np.mgrid[-n:n+1, -n:n+1]
```

```python
# Pure divergence of unit magnitude
udiv,vdiv = 1*x, 1*y

# Pure vorticity of unit magnitude
uvor,vvor = -1*y, 1*x

# Pure def1 of unit magnitude
udef1,vdef1 = 1*x, -1*y

# Pure def2 of unit magnitude
udef2,vdef2 = 1*y, 1*x

# Pure U and V
uU, vU = 1, 0
uV, vV = 0, 1
```

```python
# Define the display field udisp,vdisp and give it a label. 
# Edit-> Copy and paste this cell, then adjust the top 2 lines for the other cases

#Below is UNIT DIVERGENCE
udisp,vdisp = udiv, vdiv
label = 'Pure unit divergence U=0'
# Display the vector field as arrows, CENTERED ON the point they apply to
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()
# 
#Below is UNIT VORTICITY
udisp,vdisp = uvor, vvor
label = 'Pure unit vorticity U=0'
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()
#
#Below is DEFORMATION1
udisp,vdisp = udef1, vdef1
label = 'Pure unit deformation1 U=0'
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()
#
#Below is DEFORMATION2
udisp,vdisp = udef2, vdef2
label = 'Pure unit deformation2 U=0'
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()
#
```

## Answer to Exercise 4: 

Vorticity plus straight flow can be described as curvature vorticity. 

At lower levels, you have wind speeds increasing when moving away from the center point of the trough during positive vorticity which in this case we do because of the counterclockwise rotation of the vorticity field.

At upper levels, a counterclockwise curvature in the windflow occurs in trough and shortwaves which can lead to a south to north movement of air. Troughs are typically associated with low pressure systems.

## Answer to Exercise 5: 

When you combine divergence and vorticity you get a vector field that shows a vortex and then predominant divergence outside of the vortex as the vectors increase in magnitude. There is positive vorticity as the vector field is have counterclockwise curvature. Away from the 
vortex you begin to see a change from curvature to shear, more diffluence and stretching occurring. The positive vorticity displayed translates meteorologically to a trough where low-pressure systems are located. Interestingly, when many people think of low-pressure systems they think of hurricanes. While the vorticity and divergence vector field resembles the appearance of the winds in a hurricane, a hurricane has its strongest winds in its core. This vector field shows large upper level winds that would introduce warmer temperatures into a hurricane aloft and therefore ruining its structure.
